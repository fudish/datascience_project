### Considerazioni

Possiamo notare che abbiano diversi valori ".." presenti nelle colonne degli anni, per indicare valori nulli. Ai fini del calcolo e della pulizia dati, ritengo sia utile convertire i valori ".." in NaN (Not a Number) e convertire i tipi di dati da object a float nel caso dei numeri percentuali interi e decimali.

## Pulizia dei dati

Identifico colonne anni sfruttando la porzione di sigla YR

```python
# importo numpy per gestire valori nan
import numpy as np

# Identifico le colonne degli anni, usando le porzioni di sigle "YR"
year_columns = [col for col in df.columns if '[YR' in col]

print(f"Colonne degli anni identificate: {year_columns}")
```

Sostituisco valori ".." con NaN (Not a Number)

```python
for col in year_columns:
    df[col] = df[col].replace('..', np.nan)
```

Ricontrollo prime 10 righe del dataset

```python
display(df.head(20))
```

I valori NaN sono stati correttamente inseriti

### Converto tutti i tipi di dato object presenti nelle colonne anni, in tipo di dati float

```python
for col in year_columns:
    # Usiamo pd.to_numeric con errors='coerce' per convertire i valori.
    # errors='coerce' trasformerà qualsiasi valore che non può essere convertito (come NaN) in NaN.
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
display(df.info())
```

## 4 selezioni dei dati motivate (commento per ciascuna), 2 delle quali usanto .loc[] e .iloc[].

### 1 Selezione di tutti i paesi che hanno avuto fornitura elettrica minore del 30% nel 2016, usando il comando .loc

```python
# Seleziona le righe dove il valore nella colonna '2016 [YR2016]' è minore di 30.
# Usiamo .notna() per escludere le righe con valori NaN se ci sono.

low_access_countries_2016_loc = df.loc[df['2016 [YR2016]'].notna() & (df['2016 [YR2016]'] < 30)]

# Visualizza il risultato
display(low_access_countries_2016_loc)
```

Possiamo notare quanto il dato si verifichi maggiornmente nei paesi Africani

### 2 identifichiamo quali paesi hanno avuto sempre la corrente elettrica, al 100% in tutti gli anni rilevati

```python
# Crea una condizione booleana che è True solo se TUTTE le colonne degli anni sono uguali a 100.0 per una riga.
# Usiamo .all(axis=1) per verificare che la condizione sia vera per tutte le colonne specificate (asse 1).
# Escludiamo i valori NaN prima di fare il confronto, anche se per 100% non dovrebbero esserci se i dati sono corretti.
condition_100_percent_all_years = (df[year_columns] == 100.0).all(axis=1)

# Usiamo .loc[] per selezionare le righe che soddisfano questa condizione.
high_access_countries_all_years_loc = df.loc[condition_100_percent_all_years]

# risultato
display(high_access_countries_all_years_loc)
```

Possiamo notare quanti pochi continenti e paesi, rispetto al resto del mondo, hanno corrente continua, Europa, USA, Australia e qualche parte dell'Asia.

# 1 nuova colonna creata usando apply e lambda (commento)

## Aggiungo una nuova colonna al dataset, che assegna livelli high, medium,low e no data, in base a soglie predefinite. Utile a colpo d'occhio per vedere quali sono i valori alti, medi e bassi.

```python
# Aggiungo una nuova colonna 'Access_Level_2016' usando apply e lambda.
# Classifico l'accesso all'elettricità nel 2016 in base a soglie predefinite
df['Access_Level_2016'] = df['2016 [YR2016]'].apply(lambda x: 'High' if x >= 90 else ('Medium' if x >= 50 else ('Low' if x > 0 else 'No Data')))

# Visualizzo le prime 15 righe del DataFrame con la nuova colonna
display(df[['Country Name', '2016 [YR2016]', 'Access_Level_2016']].head(15))
```

## Importazione delle librerie necessarie per la visualizzazione (Matplotlib e Seaborn).

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Boxplot
Generazione di un boxplot per visualizzare la distribuzione dell'accesso all'elettricità nel 2016 per continente, con titolo, etichette e interpretazione.


```python
plt.figure(figsize=(14, 7)) # dimensione del grafico
sns.boxplot(x='Continent', y='2016 [YR2016]', data=df, palette='viridis')
plt.title('Distribuzione dell\'Accesso all\'Elettricità nel 2016 per Continente') # Titolo
plt.xlabel('Continente') # Etichetta asse x
plt.ylabel('Accesso all\'Elettricità (% della popolazione)') # Etichetta asse y
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("Interpretazione:")
print("Il boxplot mostra la distribuzione dell'accesso all'elettricità nei diversi continenti nel 2016.")
print("Continenti come Europa, Nord America e Oceania mostrano generalmente elevate percentuali mediane di accesso con intervalli interquartili ridotti, indicando un accesso elevato e costante.")
print("Asia e Sud America presentano una dispersione maggiore, suggerendo una maggiore variabilità nei livelli di accesso all'interno di questi continenti.")
print("L'Africa mostra la mediana di accesso più bassa e la dispersione maggiore, indicando significative disparità e un accesso complessivamente inferiore rispetto agli altri continenti.")
print("I 'baffi' (whiskers) dei boxplot indicano l'estensione dei dati, mentre i punti oltre i baffi (se presenti) rappresentano potenziali outlier.")
```

### Grafico a Barre: Accesso all'Elettricità per Continente nel 2012

Questo grafico a barre mostra la percentuale di accesso all'elettricità per ogni continente nel 2012, ordinata in ordine decrescente per facilitare il confronto e l'identificazione dei continenti con i livelli di accesso più alti e più bassi.

```python
# Calcola l'accesso medio all'elettricità per continente nel 2012
average_access_by_continent_2012 = df.groupby('Continent')['2012 [YR2012]'].mean().reset_index()

# Ordina i continenti per accesso medio in ordine decrescente
average_access_by_continent_2012 = average_access_by_continent_2012.sort_values(by='2012 [YR2012]', ascending=False)

plt.figure(figsize=(12, 6)) # Dimensione del grafico
sns.barplot(x='Continent', y='2012 [YR2012]', data=average_access_by_continent_2012, palette='viridis')
plt.title('Accesso Medio all\'Elettricità per Continente nel 2012') # Titolo in italiano
plt.xlabel('Continente') # Etichetta asse x in italiano
plt.ylabel('Accesso Medio all\'Elettricità (% della popolazione)') # Etichetta asse y in italiano
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("Interpretazione:")
print("Questo grafico a barre mostra l'accesso medio all'elettricità per i diversi continenti nel 2012.")
print("Si possono osservare le differenze significative nell'accesso medio tra i continenti.")
print("Continenti come l'Europa mostrano un accesso medio molto elevato, mentre altri come l'Africa hanno un accesso medio considerevolmente inferiore.")
print("Questo grafico fornisce una visione aggregata dell'accesso all'elettricità a livello continentale nel 2012.")
```

## Istogramma

Generazione di un istogramma per mostrare la distribuzione delle percentuali di accesso all'elettricità dal 2012 al 2016 nel Continente Africa, con titolo, etichette e interpretazione.


```python
# Filtra il dataframe per includere solo il continente Africa
df_africa = df[df['Continent'] == 'Africa'].copy()

# Melt il dataframe per avere gli anni in una singola colonna per l'istogramma
df_africa_melted = df_africa.melt(
    id_vars=['Country Name', 'Continent'],
    value_vars=year_columns,
    var_name='Year',
    value_name='Access Percentage'
)

# Rimuovi i valori NaN per l'istogramma
df_africa_melted.dropna(subset=['Access Percentage'], inplace=True)

plt.figure(figsize=(10, 6))
# Crea un istogramma della distribuzione delle percentuali di accesso in Africa
sns.histplot(data=df_africa_melted, x='Access Percentage', kde=True, bins=20, color='coral') # Usa la colonna Access Percentage e definisci i bin
plt.title('Distribuzione dell\'Accesso all\'Elettricità in Africa (2012-2016)') # Titolo in italiano
plt.xlabel('Accesso all\'Elettricità (% della popolazione)') # Etichetta asse x in italiano
plt.ylabel('Frequenza') # Etichetta asse y in italiano
plt.grid(axis='y', alpha=0.75)
plt.show()

print("Interpretazione:")
print("Questo istogramma mostra la distribuzione di frequenza delle percentuali di accesso all'elettricità per i paesi in Africa nel periodo 2012-2016.")
print("La maggior parte delle osservazioni si concentra nelle fasce di accesso più basse, indicando che molti paesi africani hanno ancora bassi livelli di accesso all'elettricità.")
print("La coda verso destra indica che ci sono anche paesi africani con livelli di accesso più elevati, ma sono meno numerosi.")
print("Questo grafico evidenzia le significative sfide che molti paesi africani affrontano nel fornire accesso universale all'elettricità.")
```

## Scatterplot

Generazione di uno scatterplot per esplorare la relazione tra l'accesso all'elettricità nel 2012 e nel 2016, con titolo, etichette e interpretazione.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='2012 [YR2012]', y='2016 [YR2016]', hue='Continent', palette='viridis', s=50) # Aggiunto hue, palette e dimensione punti
plt.title('Accesso all\'Elettricità: 2012 vs 2016') # Titolo in italiano
plt.xlabel('Accesso all\'Elettricità nel 2012 (%)') # Etichetta asse x in italiano
plt.ylabel('Accesso all\'Elettricità nel 2016 (%)') # Etichetta asse y in italiano
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Continente') # Aggiunta legenda
plt.show()

print("Interpretazione:")
print("Lo scatter plot mostra la relazione tra l'accesso all'elettricità nel 2012 e nel 2016 per diversi paesi e regioni.")
print("Esiste una forte correlazione positiva tra i due anni, indicando che i paesi con accesso più elevato nel 2012 generalmente avevano anche un accesso più elevato nel 2016.")
print("La maggior parte dei punti dati si trova vicino alla linea diagonale (che rappresenta un accesso uguale in entrambi gli anni), suggerendo che i livelli di accesso sono rimasti relativamente stabili per molti paesi.")
print("Alcuni punti sopra la diagonale indicano paesi in cui l'accesso all'elettricità è aumentato dal 2012 al 2016.")
print("Al contrario, i punti al di sotto della diagonale (anche se in numero inferiore) suggeriscono una diminuzione dell'accesso per quelle entità.")
print("Il raggruppamento di punti vicino al 100% indica che molti paesi avevano già un accesso completo o quasi completo in entrambi gli anni.")
print("L'uso dei colori per i continenti evidenzia visivamente come i paesi di diversi continenti si raggruppino in diverse aree del grafico, mostrando le disparità regionali nell'accesso all'elettricità.")
```

## Barplot

Generazione di un grafico barplot per i 10 paesi con accesso più alto, con titolo, etichette e interpretazione.


```python
top_10_countries_2016 = df.dropna(subset=['2016 [YR2016]']).sort_values(by='2016 [YR2016]', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Country Name', y='2016 [YR2016]', data=top_10_countries_2016, palette='viridis')
plt.title('Primi 10 Paesi per Accesso all\'Elettricità nel 2016') # Titolo
plt.xlabel('Nome del Paese') # Etichetta asse x in italiano
plt.ylabel('Accesso all\'Elettricità (% della popolazione)') # Etichetta asse y
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("Interpretazione del grafico a barre:")
print("Questo grafico a barre mostra i primi 10 paesi con la più alta percentuale di popolazione con accesso all'elettricità nel 2016.")
print("Come previsto, questi paesi mostrano tassi di accesso molto vicini o pari al 100%, indicando un accesso quasi universale.")
print("Questa visualizzazione evidenzia chiaramente i paesi che hanno raggiunto un successo significativo nel fornire elettricità alle loro popolazioni entro il 2016.")
```
